```python
import pandas as pd
from plotly.offline import init_notebook_mode, iplot
from plotly.graph_objs import *

init_notebook_mode(connected=True)         # initiate notebook for offline plot

df = pd.read_csv("./output/fort.21_1_1", sep='\s+', header=None).values
dfr = pd.read_csv("./output/fort.22_1_1", sep='\s+', header=None).values
dfxp = pd.read_csv("./output/fort.23_1_1", sep='\s+', header=None).values


n = 128                                    # number of coordinate steps
nt = 64                                    # maximum number of time steps
mod = 1                                    # to show each mod-th time point if needed 
trace = []
xf = []
yf = []
tf = []
for l in range(nt): 
    if not l%mod:
        xf.append(df[l*n:(l + 1)*n, 1])
        yf.append(df[l*n:(l + 1)*n, 2])
        tf.append(df[l*n:l*n + 1, 0])

for l in range(len(xf)):
    trace.append(Scatter(x = xf[l],y = yf[l],name = str(tf[l]).strip('[]')))
    
data = Data(trace)
fig = Figure(data)

fig.update_layout(
    title={
        'text': "Absolute value of the wave function",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis_title={
        'text': 'coordinate'
    },
    yaxis_title={
        'text': 'abs()'
    }
)

#iplot(fig)               # use plotly.offline.iplot for offline plot
iplot(fig, image='svg', filename='fig_abs_wf', image_width=1280, image_height=780)

tracer = []
xfr = []
yfr = []
tfr = []
for l in range(nt): 
    if not l%mod:
        xfr.append(dfr[l*n:(l + 1)*n, 1])
        yfr.append(dfr[l*n:(l + 1)*n, 2])
        tfr.append(dfr[l*n:l*n + 1, 0])

for l in range(len(xfr)):
    tracer.append(Scatter(x = xfr[l],y = yfr[l],name = str(tfr[l]).strip('[]')))
    
datar = Data(tracer)
figr = Figure(datar)

figr.update_layout(
    title={
        'text': "Real value of the wave function",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis_title={
        'text': 'coordinate'
    },
    yaxis_title={
        'text': 'Re()'
    }
)
#iplot(figr)    # use plotly.offline.iplot for offline plot
iplot(figr, image='svg', filename='fig_real_wf', image_width=1280, image_height=780)

tracem = []

namem = ["<x>", "<x^2>", "<p>", "<p^2>"]
    
for l in range(1,5):
    tracem.append(Scatter(x = dfxp[:, 0],y = dfxp[:, l],name = namem[l-2]))
    
datam = Data(tracem)
figm = Figure(datam)

figm.update_layout(
    title={
        'text': "Expectation values",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis_title={
        'text': 'time'
    },
    yaxis_title={
        'text': '<A>'
    }
)

#iplot(figm)    # use plotly.offline.iplot for offline plot
iplot(figm, image='svg', filename='fig_real_m', image_width=1280, image_height=780)
```
